```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# Download necessary NLTK data
# nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
```

```python
def remove_stopword(text):
    # text = 'An unusually intense squall line, feeding on a...'
    word_tokens = word_tokenize(text)

    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]

    filtered_text = " ".join(filtered_sentence)
    return filtered_text
```

```python
# Apply the remove_stopword function to the 'EPISODE_NARRATIVE' column
df['cleaned_episode_narrative'] = df['EPISODE_NARRATIVE'].apply(str).apply(remove_stopword)

# Display the first few rows with the original and cleaned narratives
display(df[['EPISODE_NARRATIVE', 'cleaned_episode_narrative']].head())
```

```python
from nltk import pos_tag
# Download required NLTK data
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# Tokenize and POS tag each row
def pos_tag_text(text):
    tokens = word_tokenize(text)
    tagged = pos_tag(tokens)
    return tagged

df['pos_tags'] = df['cleaned_episode_narrative'].apply(pos_tag_text)
```

```python
from nltk.stem import PorterStemmer

# Initialize the Porter Stemmer
ps = PorterStemmer()

def stem_words(sentence):
    """Applies stemming to a list of tokens."""
    return [ps.stem(word) for word in sentence.split()]

# Apply stemming to the 'cleaned_narrative' column
df['stemmed_narrative'] = df['cleaned_episode_narrative'].apply(stem_words)

# Display the first few rows with the cleaned and stemmed narratives
display(df[['cleaned_episode_narrative', 'stemmed_narrative']].head())
```

```python
from gensim.models import Word2Vec

# Train a Word2Vec model on the stemmed narratives
# min_count: Ignores all words with total frequency lower than this.
# vector_size: Dimensionality of the word vectors.
# workers: Use these many worker threads to train the model (=faster training with multicore CPUs).
model = Word2Vec(sentences=df['stemmed_narrative'],
                 vector_size=100,  # size of word vectors
                 window=5,         # context window size
                 min_count=1,      # ignore words with total frequency lower than this
                 workers=4,        # parallelism
                 sg=1              # skip-gram (1), cbow (0)
                )

# You can access the vector for a specific word like this:
# print(model.wv['flood'])

# Find similar words:
# similar_words = model.wv.most_similar('learning')
# print(similar_words)

# Check similarity between two words:
# similarity = model.wv.similarity('machine', 'learning')
# print(similarity)

# To get sentence embeddings, you could average the word vectors for each sentence.
# For example:
def get_sentence_vector(word_list):
    vectors = [model.wv[word] for word in word_list if word in model.wv]
    if not vectors:
        return None
    return sum(vectors) / len(vectors)

df['narrative_vector'] = df['stemmed_narrative'].apply(get_sentence_vector)

# Display the first few rows with the new vector column (will show arrays)
display(df[['stemmed_narrative', 'narrative_vector']].head())
```
